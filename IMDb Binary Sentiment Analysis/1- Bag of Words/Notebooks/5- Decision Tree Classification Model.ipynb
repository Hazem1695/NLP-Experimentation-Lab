# **Data preprocessing**

## Data Cleaning Check Template
This template is designed to quickly assess the quality of any dataset before building machine learning models or performing analysis.

It provides a structured overview of the dataset by checking for common data issues such as:

- Missing values

- Duplicate rows

- Incorrect data types

- Outliers

- Distribution of numerical features

- Categorical feature consistency

**What This Template Does**

- Displays basic dataset information (shape, data types)

- Identifies missing values and duplicates

- Summarizes numerical and categorical features

- Detects potential outliers using the IQR method

- Highlights columns with low unique values for quick inspection

How to Use

1. Load your dataset using Pandas  

2. Call the function:

````python
def data_quality_report(df):

    print("DATA QUALITY REPORT")
    
    # Print a separator line for better readability
    
    print("=" * 50)
    print("BASIC INFO")
    print("=" * 50)
    
    # Show general information about the dataset (columns, data types, non-null values)
    print(df.info())
    
    # Show number of rows and columns
    print("\n" + "=" * 50)
    print("SHAPE OF DATA")
    print("=" * 50)
    print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
    
    # Check for missing (null) values in each column
    print("\n" + "=" * 50)
    print("MISSING VALUES")
    print("=" * 50)
    missing = df.isnull().sum()
    
    # Display only columns that have missing values
    print(missing[missing > 0])
    
    # Check for duplicate rows
    print("\n" + "=" * 50)
    print("DUPLICATES")
    print("=" * 50)
    print(f"Duplicate rows: {df.duplicated().sum()}")
    
    # Display data types of each column
    print("\n" + "=" * 50)
    print("DATA TYPES")
    print("=" * 50)
    print(df.dtypes)
    
    # Summary statistics for numerical columns (mean, std, min, max, etc.)
    print("\n" + "=" * 50)
    print("NUMERICAL SUMMARY")
    print("=" * 50)
    print(df.describe())
    
    # Summary for categorical (object) columns
    print("\n" + "=" * 50)
    print("CATEGORICAL SUMMARY")
    print("=" * 50)
    print(df.describe(include=['object']))
    
    # Show unique values for columns with low number of distinct values
    # Useful for detecting categories, errors, or inconsistencies
    print("\n" + "=" * 50)
    print("UNIQUE VALUES (LOW CARDINALITY)")
    print("=" * 50)
    for col in df.columns:
        if df[col].nunique() < 10:  # Only show columns with few unique values
            print(f"{col}: {df[col].unique()}")
            
    # correlation
    print("\n" + "=" * 50)
    print("CORRELATION MATRIX")
    print("=" * 50)
    print(df.corr(numeric_only=True))
    
    # Detect outliers using the IQR (Interquartile Range) method
    print("\n" + "=" * 50)
    print("OUTLIERS CHECK (IQR METHOD)")
    print("=" * 50)
    
    # Loop through only numerical columns
    for col in df.select_dtypes(include=np.number).columns:
        Q1 = df[col].quantile(0.25)  # 25th percentile
        Q3 = df[col].quantile(0.75)  # 75th percentile
        IQR = Q3 - Q1  # Interquartile range
        
        # Count rows that fall outside the normal range
        outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
        print(f"{col}: {len(outliers)} outliers")
````

## **Cleaning the texts**

````python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
import re
import contractions
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

base_stopwords = set(stopwords.words('english'))
negation_words = {'not', 'no', 'never'}
all_stopwords = base_stopwords - negation_words

wnl = WordNetLemmatizer()

corpus = []

for i in range(0, len(dataset)):
    review = dataset['review'][i]
    # Fix contractions
    review = contractions.fix(review)
    # Lowercase
    review = review.lower()
    
    review = re.sub(r'[^a-zA-Z\s]', ' ', review)
    # Split
    words = review.split()
    # Chained Lemmatization (Handles both Verbs 'v' and Nouns 'n')
    review = [wnl.lemmatize(wnl.lemmatize(word, pos='v'), pos='n') for word in words if word not in all_stopwords]
    review = ' '.join(review) 
    corpus.append(review)
````

````python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(splitter='random', max_depth=20, min_samples_split=2, min_samples_leaf=2, criterion='gini', random_state=0)
classifier.fit(X_train, y_train)
````

# Decision Tree Performance Analysis for Text Classification

# 1. Objective

The objective of this experiment was to evaluate the effectiveness of the **Decision Tree** algorithm for binary text classification using **Bag-of-Words (BoW)** features generated by **CountVectorizer**.

Unlike linear models and probabilistic classifiers, Decision Trees partition the feature space through a sequence of hierarchical decisions. Although Decision Trees are highly interpretable, they are often challenged by high-dimensional and sparse text representations. This study investigates how feature extraction, tree complexity, and hyperparameter optimization influence their performance.

The experiment aims to answer the following research questions:

* How well does Decision Tree perform on sparse text data?
* Does increasing the vocabulary size always improve performance?
* Can hyperparameter optimization significantly improve the classifier?
* Why was **RandomizedSearchCV** selected instead of **GridSearchCV**?
* What are the limitations of Decision Trees for text classification?

# 3. Hyperparameter Optimization Strategy

**GridSearchCV** was considered first for hyperparameter optimization.

The complete search space contained:

* **1,728 parameter combinations**
* **5-fold Cross Validation**

Resulting in:

```text
1,728 × 5 = 8,640 model trainings
```

Training thousands of Decision Trees on high-dimensional text data would require a considerable amount of computation time.

Instead, **RandomizedSearchCV** was selected.

Configuration:

```text
30 random parameter combinations
5-fold Cross Validation
```

Resulting in only:

```text
30 × 5 = 150 model trainings
```

This reduced the computational cost by approximately **98%** (150 vs. 8,640 fits) while exploring only **~1.7%** of the full grid — RandomizedSearchCV was chosen because it is substantially faster than GridSearchCV while still surfacing strong configurations.

---

# 4. Hyperparameters Explored

The optimization process searched over parameters including:

* Tree depth (`max_depth`)
* Splitting strategy (`splitter`)
* Splitting criterion (`gini`, `entropy`)
* Minimum samples required for splitting (`min_samples_split`)
* Minimum samples required per leaf (`min_samples_leaf`)
* Feature selection strategy (`max_features`)

**Notably, RandomizedSearchCV converged on the exact same best configuration for all three vocabulary sizes (5K, 10K, 15K):**

```python
DecisionTreeClassifier(
    splitter='random',
    criterion='gini',
    max_depth=20,
    min_samples_split=2,
    min_samples_leaf=2,
    max_features=None
)
```

This consistency suggests the search space has a stable optimum for this dataset that is largely independent of vocabulary size — the tree's own capacity constraints (depth, leaf/split minimums), rather than the number of available features, appear to be the dominant factor in performance.

# 6. Performance Analysis

## Effect of Hyperparameter Optimization

Hyperparameter tuning produced a clear improvement over the untuned baseline.

Accuracy increased from:

**71.37%** (baseline)

to

**74.34%** (best tuned model, 5K features)

representing an improvement of nearly **3 percentage points**, along with a tighter, more consistent cross-validation spread.

## Why RandomizedSearchCV?

| Method             | Candidates | Total Fits |
| ------------------- | ---------: | ---------: |
| GridSearchCV        |      1,728 |      8,640 |
| RandomizedSearchCV  |         30 |        150 |

RandomizedSearchCV evaluated only **1.7%** of the total parameter combinations yet discovered a configuration that meaningfully outperformed the untuned baseline, at roughly **2% of the computational cost** of an exhaustive grid search.

## Effect of Vocabulary Size

| Features | Accuracy   |
| -------- | ---------- |
| **5K**   | **74.34%** |
| 10K      | 74.08%     |

Unlike in prior BoW experiments (e.g. Naive Bayes, LinearSVC), where accuracy tended to improve or plateau as vocabulary size increased, **the 5K-feature configuration slightly outperformed the 10K-feature configuration here (74.34% vs. 74.08%)**. This is a notable reversal of the "more features → better" pattern.

A likely explanation is that a single Decision Tree's capacity is bounded by its depth and leaf constraints (`max_depth=20`, `min_samples_leaf=2`) rather than by the size of the input vocabulary — beyond a certain point, additional features mostly add sparse, rarely-used splitting candidates that dilute the search rather than improve it. This reinforces the broader conclusion (Section 12) that Decision Trees do not scale their accuracy with vocabulary size the way linear models do.

## Effect of Splitting Strategy

The optimized models consistently selected:

```python
splitter='random'
```

rather than `splitter='best'`. Random splitting introduces additional randomness during tree construction, which reduces overfitting and tends to generalize better on high-dimensional sparse data, where greedily choosing the locally optimal split can produce overly specific decision boundaries.

# 12. Discussion

The experiments confirm the general limitation of a single Decision Tree for sparse text classification: axis-aligned splits struggle to build compact, generalizable boundaries over thousands of mostly-zero BoW dimensions.

RandomizedSearchCV again proved to be a practical substitute for GridSearchCV, cutting required trainings from 8,640 to 150 (~98% reduction) while still finding a configuration that improved accuracy by ~3 points over the untuned baseline.

The most notable result in this run is that **increasing vocabulary size did not help** — 5K features slightly outperformed and 10K. Since the same tree hyperparameters were selected regardless of vocabulary size, the bottleneck appears to be the tree's structural capacity (depth and leaf constraints) rather than the richness of the feature set. This is a useful contrast to linear/probabilistic models, where larger vocabularies more reliably translate into better performance.

---

# 13. Final Conclusion

This experiment evaluated Decision Tree performance on binary text classification using Bag-of-Words features at three vocabulary sizes, with RandomizedSearchCV used for hyperparameter tuning.

The best-evaluated model used **CountVectorizer with 5,000 features and bigrams**, paired with an optimized **DecisionTreeClassifier** (random splitting, Gini impurity, max_depth=20), achieving **74.34% accuracy** and **0.7441 ROC-AUC** — outperforming the untuned baseline by nearly 3 points and edging out the 10K-feature configuration.

Unlike prior experiments, more features did not translate into better performance here, and the 10K configuration showed the more stable cross-validation spread despite a marginally lower mean accuracy. Overall, these results reinforce that single Decision Trees are capacity-bound rather than vocabulary-bound, and remain a weaker choice for high-dimensional sparse text data compared to linear and probabilistic models.
