# Final Experimental Report

# IMDb Binary Sentiment Analysis using Bag of Words

---

# Executive Summary

This study presents a comprehensive experimental evaluation of classical machine learning algorithms for **binary sentiment classification** on the IMDb movie review dataset using the **Bag of Words (BoW)** text representation.

Rather than evaluating a single classifier, this project systematically investigates multiple supervised learning algorithms under different feature extraction configurations and hyperparameter settings to determine which approach is most suitable for sparse high-dimensional textual data.

The experiments cover the complete machine learning workflow, including feature engineering, model optimization, cross-validation, performance evaluation, computational analysis, and comparative benchmarking.

More than simply reporting accuracy, this work focuses on understanding **why certain algorithms perform better than others when applied to Bag of Words features.**

---

# Objectives

The primary objectives of this experimental study were to:

* Investigate the effectiveness of Bag of Words for sentiment analysis.
* Compare multiple machine learning algorithms under identical experimental conditions.
* Study the influence of vocabulary size on classification performance.
* Evaluate different hyperparameter optimization strategies.
* Analyze computational efficiency versus predictive performance.
* Identify the most suitable classifier for sparse text representations.

---

# Dataset

**Dataset**

[IMDb Binary Sentiment Analysis](https://www.kaggle.com/datasets/mwallerphunware/imbd-movie-reviews-for-binary-sentiment-analysis)

Binary Text Classification

**Classes**

* Positive Review
* Negative Review

# Hyperparameter Optimization

Different optimization strategies were employed depending on the computational requirements of each algorithm.

Techniques used include:

* Manual tuning
* RandomizedSearchCV
* HalvingRandomSearchCV

GridSearchCV was intentionally avoided for several experiments because the computational cost was prohibitively high.

For example:

Decision Tree

GridSearchCV

* 1,728 parameter combinations
* 5-fold cross validation
* **8,640 model fits**

RandomizedSearchCV

* 30 random combinations
* 5-fold cross validation
* **150 model fits**

The randomized approach dramatically reduced computational time while still identifying high-quality hyperparameters.

Similarly, HalvingRandomSearchCV was selected for Random Forest because it provides efficient successive resource allocation and significantly faster convergence compared to exhaustive search.

# Experimental Results

## Logistic Regression

Logistic Regression consistently delivered one of the strongest performances across all experiments.

Key observations:

* Performance improved as vocabulary size increased.
* Stable cross-validation scores.
* Excellent balance between precision and recall.
* Low variance across folds.
* Fast training.
* Highly scalable.

Best configuration

* 30,000 features
* Accuracy: **88.92%**
* ROC-AUC: **0.8894**

This model established itself as one of the strongest baselines for the entire project.

---

## Linear Support Vector Machine

LinearSVC demonstrated outstanding performance on sparse Bag of Words representations.

Observations:

* Training was significantly faster than kernel SVM.
* Linear decision boundaries generalized well.
* Excellent scalability.
* Very stable cross-validation performance.

Best configuration

* 30,000 features
* Accuracy: **87.71%**
* ROC-AUC: **0.8773**

Attempts to use kernel SVM were abandoned because training time became prohibitively expensive on high-dimensional text data.

---

## Random Forest

Random Forest showed substantial improvement after hyperparameter optimization.

Important findings:

* HalvingRandomSearchCV substantially accelerated optimization.
* Increasing vocabulary improved performance.
* Ensemble learning reduced overfitting compared with Decision Trees.
* Balanced precision and recall.

Best configuration

* 15,000 features
* Accuracy: **85.83%**
* ROC-AUC: **0.8586**

Random Forest proved considerably stronger than a single Decision Tree but remained behind linear models.

---

## Decision Tree

Decision Trees exhibited limited performance on sparse text features.

Observations:

* Sensitive to hyperparameters.
* Improved after RandomizedSearchCV.
* Higher variance.
* Less effective for sparse BoW vectors.

Best configuration

* 10,000–15,000 features
* Accuracy: approximately **74%**

Although interpretable, Decision Trees were unable to compete with linear classifiers.

---

## K-Nearest Neighbors

KNN produced the weakest overall performance.

Major findings:

* Euclidean distance performed poorly in sparse spaces.
* Cosine similarity substantially improved results.
* Computationally expensive during prediction.
* Performance remained below linear classifiers.

Best configuration

* Cosine distance
* k = 11
* Accuracy: **75.31%**

The experiments highlight why KNN is generally not preferred for high-dimensional text classification.

---

## Gaussian Naive Bayes

Gaussian Naive Bayes struggled because it assumes continuous Gaussian-distributed features.

Observations:

* Lower accuracy.
* Lower recall.
* Inferior fit for count-based representations.

Best configuration

Accuracy approximately **81.7%**

The Gaussian assumption is not appropriate for sparse word-count vectors.

---

## Multinomial Naive Bayes

Multinomial Naive Bayes performed exceptionally well.

Observations:

* Naturally suited for word counts.
* Extremely fast.
* Excellent baseline.
* Minimal computational cost.

Best configuration

Accuracy approximately **87.56%**

This model remains one of the strongest classical approaches for Bag of Words text classification.

---

## Bernoulli Naive Bayes

Bernoulli Naive Bayes also achieved competitive performance.

Observations:

* Effective with binary word occurrence.
* Stable training.
* Very efficient.

Best configuration

Accuracy approximately **87.66%**

Performance was comparable to Multinomial Naive Bayes.

# Key Findings

Several important conclusions emerged from these experiments:

* Increasing vocabulary size consistently improved predictive performance.
* Sparse linear models outperform distance-based methods.
* Multinomial and Bernoulli Naive Bayes remain remarkably competitive despite their simplicity.
* Logistic Regression achieved the highest overall performance.
* LinearSVC provides an excellent balance between accuracy and computational efficiency.
* Random Forest benefits significantly from hyperparameter optimization but cannot fully exploit sparse textual representations.
* Decision Trees are prone to instability on high-dimensional Bag of Words data.
* Cosine similarity is substantially better than Euclidean distance for KNN in text classification.
* Efficient hyperparameter optimization methods greatly reduce computational cost without sacrificing model quality.

---

# Computational Observations

Several computational challenges were encountered throughout the experiments.

* Full CountVectorizer occasionally exhausted available RAM.
* Kernel SVM required prohibitively long training times.
* GridSearchCV became impractical due to the exponential growth of parameter combinations.
* RandomizedSearchCV provided an excellent balance between speed and optimization quality.
* HalvingRandomSearchCV significantly accelerated Random Forest tuning.

These observations reinforce the importance of selecting algorithms that scale effectively to sparse high-dimensional feature spaces.

---

# Final Conclusion

This experimental study demonstrates that **Bag of Words remains a powerful feature representation for classical NLP tasks** when combined with appropriate machine learning algorithms.

Among all evaluated models, **Logistic Regression** consistently achieved the strongest overall performance, offering the best combination of predictive accuracy, stability, computational efficiency, and scalability.

LinearSVC closely followed while providing exceptional training efficiency for high-dimensional sparse data.

Multinomial and Bernoulli Naive Bayes proved that simple probabilistic models remain highly competitive for document classification.

Although Random Forest improved substantially after optimization, ensemble tree methods were still unable to surpass linear models specifically designed for sparse feature spaces.

Finally, the study highlights that successful NLP systems depend not only on choosing the right classifier but also on careful feature engineering, systematic experimentation, rigorous evaluation, and computationally efficient optimization strategies.
