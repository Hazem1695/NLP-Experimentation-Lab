# **Data preprocessing**

## Data Cleaning Check Template
This template is designed to quickly assess the quality of any dataset before building machine learning models or performing analysis.

It provides a structured overview of the dataset by checking for common data issues such as:

- Missing values

- Duplicate rows

- Incorrect data types

- Outliers

- Distribution of numerical features

- Categorical feature consistency

**What This Template Does**

- Displays basic dataset information (shape, data types)

- Identifies missing values and duplicates

- Summarizes numerical and categorical features

- Detects potential outliers using the IQR method

- Highlights columns with low unique values for quick inspection

How to Use

1. Load your dataset using Pandas  

2. Call the function:

````python
def data_quality_report(df):

    print("DATA QUALITY REPORT")
    
    # Print a separator line for better readability
    
    print("=" * 50)
    print("BASIC INFO")
    print("=" * 50)
    
    # Show general information about the dataset (columns, data types, non-null values)
    print(df.info())
    
    # Show number of rows and columns
    print("\n" + "=" * 50)
    print("SHAPE OF DATA")
    print("=" * 50)
    print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
    
    # Check for missing (null) values in each column
    print("\n" + "=" * 50)
    print("MISSING VALUES")
    print("=" * 50)
    missing = df.isnull().sum()
    
    # Display only columns that have missing values
    print(missing[missing > 0])
    
    # Check for duplicate rows
    print("\n" + "=" * 50)
    print("DUPLICATES")
    print("=" * 50)
    print(f"Duplicate rows: {df.duplicated().sum()}")
    
    # Display data types of each column
    print("\n" + "=" * 50)
    print("DATA TYPES")
    print("=" * 50)
    print(df.dtypes)
    
    # Summary statistics for numerical columns (mean, std, min, max, etc.)
    print("\n" + "=" * 50)
    print("NUMERICAL SUMMARY")
    print("=" * 50)
    print(df.describe())
    
    # Summary for categorical (object) columns
    print("\n" + "=" * 50)
    print("CATEGORICAL SUMMARY")
    print("=" * 50)
    print(df.describe(include=['object']))
    
    # Show unique values for columns with low number of distinct values
    # Useful for detecting categories, errors, or inconsistencies
    print("\n" + "=" * 50)
    print("UNIQUE VALUES (LOW CARDINALITY)")
    print("=" * 50)
    for col in df.columns:
        if df[col].nunique() < 10:  # Only show columns with few unique values
            print(f"{col}: {df[col].unique()}")
            
    # correlation
    print("\n" + "=" * 50)
    print("CORRELATION MATRIX")
    print("=" * 50)
    print(df.corr(numeric_only=True))
    
    # Detect outliers using the IQR (Interquartile Range) method
    print("\n" + "=" * 50)
    print("OUTLIERS CHECK (IQR METHOD)")
    print("=" * 50)
    
    # Loop through only numerical columns
    for col in df.select_dtypes(include=np.number).columns:
        Q1 = df[col].quantile(0.25)  # 25th percentile
        Q3 = df[col].quantile(0.75)  # 75th percentile
        IQR = Q3 - Q1  # Interquartile range
        
        # Count rows that fall outside the normal range
        outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
        print(f"{col}: {len(outliers)} outliers")
````

## **Cleaning the texts**

````python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
import re
import contractions
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

base_stopwords = set(stopwords.words('english'))
negation_words = {'not', 'no', 'never'}
all_stopwords = base_stopwords - negation_words

wnl = WordNetLemmatizer()

corpus = []

for i in range(0, len(dataset)):
    review = dataset['review'][i]
    # Fix contractions
    review = contractions.fix(review)
    # Lowercase
    review = review.lower()
    
    review = re.sub(r'[^a-zA-Z\s]', ' ', review)
    # Split
    words = review.split()
    # Chained Lemmatization (Handles both Verbs 'v' and Nouns 'n')
    review = [wnl.lemmatize(wnl.lemmatize(word, pos='v'), pos='n') for word in words if word not in all_stopwords]
    review = ' '.join(review) 
    corpus.append(review)
````

## Preprocessing Verification
> **Note:** Pulling the first two rows directly as a memory array to confirm that our lowercasing, stopword stripping, and lemmatization pipeline worked correctly before feeding it into the vectorizer.

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(corpus, y, test_size = 0.20, random_state = 0)
````

# **Creating the TF-IDF (Term Frequency-Inverse Document Frequency) model**

````python
from sklearn.feature_extraction.text import TfidfVectorizer
# Add ngram_range=(1, 2) so it automatically catches phrases like "not good" or "no clue"
tf = TfidfVectorizer(max_features=10000, ngram_range=(1,2))
X_train = tf.fit_transform(X_train)
X_test = tf.transform(X_test)
````

````python
print(X_train.shape)        # (n_samples, n_features)
print(X_train.shape[0])     # number of training samples
print(X_train.shape[1])     # number of features (vocabulary size)
````

````python
print(X_test.shape)        # (n_samples, n_features)
print(X_test.shape[0])     # number of training samples
print(X_test.shape[1])     # number of features (vocabulary size)
````

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score, classification_report

dt = DecisionTreeClassifier(random_state=0)

param_grid = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'splitter': ['best', 'random'],
    'max_depth': [None, 5, 10, 20, 30, 50],
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 2, 4, 8],
    'max_features': [None, 'sqrt', 'log2']
}
````

````python
random_search = RandomizedSearchCV(
    estimator=dt,
    param_distributions=param_grid,
    n_iter=30,          # Number of random combinations
    scoring='accuracy',
    cv=5,
    random_state=42,
    n_jobs=-1,
    verbose=2
)
````

````python
random_search.fit(X_train, y_train)

print(random_search.best_params_)
print(random_search.best_score_)
````

# **Training the Decision Tree model on the Training set**

````python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(splitter='random', max_depth=20, min_samples_split=2, min_samples_leaf=2, criterion='gini', random_state=0)
classifier.fit(X_train, y_train)
````

# **Predicting the Test set results**

````python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
````

# **Evaluating the Model Performance**

````python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_auc_score
from sklearn.model_selection import cross_val_score

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nROC-AUC Score:")
print(roc_auc_score(y_test, y_pred))


accuracies = cross_val_score(estimator=classifier, X=X_train, y=y_train, cv=3)

print("\nMean Accuracy:")
print(accuracies.mean())

print("\nStandard Deviation:")
print(accuracies.std())
````

# 1. Objective

The objective of this experiment was to evaluate the effectiveness of the **Decision Tree** algorithm for binary text classification using features generated by **TfidfVectorizer**.

Unlike linear models and probabilistic classifiers, Decision Trees partition the feature space through a sequence of hierarchical decisions. Although Decision Trees are highly interpretable, they are often challenged by high-dimensional and sparse text representations. This study investigates how feature extraction, tree complexity, and hyperparameter optimization influence their performance under TF-IDF weighting.

The experiment aims to answer the following research questions:

* How well does Decision Tree perform on sparse, TF-IDF-weighted text data?
* Does increasing the vocabulary size improve performance?
* Can hyperparameter optimization significantly improve the classifier?
* Why was **RandomizedSearchCV** selected instead of **GridSearchCV**?
* How does the Decision Tree compare against previously evaluated TF-IDF models?

# 3. Hyperparameter Optimization Strategy

**GridSearchCV** was considered first for hyperparameter optimization.

The complete search space contained:

* **1,728 parameter combinations**
* **5-fold Cross Validation**

Resulting in:

```text
1,728 × 5 = 8,640 model trainings
```

Training thousands of Decision Trees on high-dimensional text data would require a considerable amount of computation time.

Instead, **RandomizedSearchCV** was selected.

Configuration:

```text
30 random parameter combinations
5-fold Cross Validation
```

Resulting in only:

```text
30 × 5 = 150 model trainings
```

This reduced the computational cost by approximately **98%** (150 vs. 8,640 fits) while exploring only **~1.7%** of the full grid — RandomizedSearchCV was chosen because it is substantially faster than GridSearchCV while still surfacing strong configurations.

---

* Feature selection strategy (`max_features`)

## Effect of Hyperparameter Optimization

Hyperparameter tuning produced a clear improvement over the untuned baseline.

Accuracy increased from:

**70.53%** (baseline)

to

**74.94%** (best tuned model, 10K features)

## Why RandomizedSearchCV?

| Method             | Candidates | Total Fits |
| ------------------- | ---------: | ---------: |
| GridSearchCV        |      1,728 |      8,640 |
| RandomizedSearchCV  |         30 |        150 |

RandomizedSearchCV evaluated only **1.7%** of the total parameter combinations yet discovered a configuration that meaningfully outperformed the untuned baseline, at roughly **2% of the computational cost** of an exhaustive grid search.

## Effect of Vocabulary Size

| Features | Accuracy   |
| -------- | ---------- |
| 5K       | 71.99%     |
| **10K**  | **74.94%** |
| 15K      | 73.98%     |

Accuracy increased sharply from 5K to 10K features, then declined slightly at 15K. This is a cleaner "rise then plateau/decline" pattern than what was observed in the earlier CountVectorizer-based Decision Tree experiment (where 5K unexpectedly edged out 10K). Under TF-IDF weighting, 10K appears to be the sweet spot — enough vocabulary to capture useful splits, without diluting the tree with too many low-signal, high-dimensional features.

## Effect of Splitting Strategy

The optimized models consistently selected:

```python
splitter='random'
```

rather than `splitter='best'`. Random splitting introduces additional randomness during tree construction, which reduces overfitting and tends to generalize better on high-dimensional sparse data, where greedily choosing the locally optimal split can produce overly specific decision boundaries.

# 12. Discussion

The experiments confirm the general limitation of a single Decision Tree for sparse, high-dimensional text classification, even under TF-IDF weighting: axis-aligned splits still struggle to build compact, generalizable boundaries over thousands of largely-zero dimensions.

RandomizedSearchCV again proved to be a practical substitute for GridSearchCV, cutting required trainings from 8,640 to 150 (~98% reduction) while still finding configurations that meaningfully improved accuracy over the untuned baseline.

Unlike the earlier CountVectorizer-based run — where 5K features unexpectedly outperformed 10K — this TF-IDF run showed the more intuitive pattern of accuracy rising to a peak at 10K features before declining slightly at 15K. This suggests TF-IDF's down-weighting of common terms makes the tree's splits more informative as vocabulary grows, at least up to a point, whereas raw counts (BoW) let the tree's outcome depend more on incidental noise in which of the larger feature set happened to get selected.

Across both feature extraction methods, Decision Tree remains substantially behind linear and probabilistic models, reinforcing that a single tree is not well suited to high-dimensional sparse text representations regardless of whether counts or TF-IDF weighting are used.

# 13. Final Conclusion

This experiment evaluated Decision Tree performance on binary text classification using TF-IDF features across four configurations, with RandomizedSearchCV used for hyperparameter tuning.

The best-performing model used **TfidfVectorizer with 10,000 features and bigrams**, paired with an optimized **DecisionTreeClassifier** (random splitting, Gini impurity, max_depth=20), achieving **74.94% accuracy** and **0.7504 ROC-AUC** — its strongest result across either feature extraction method tried so far.

Compared against previously evaluated TF-IDF classifiers, Decision Tree ranks last, well behind LinearSVC (90.18%), Logistic Regression (89.46%), Multinomial Naive Bayes (88.13%), and KNN (80.41%). These results reinforce that single Decision Trees, while interpretable, are not the most suitable choice for high-dimensional sparse text classification — ensemble methods such as Random Forest are the natural next step to test whether combining multiple trees closes this gap.
